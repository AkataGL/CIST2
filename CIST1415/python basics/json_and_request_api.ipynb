```python
import json
```

```python
emp_string = '''
{
   "employees":[
      {
         "name":"John Doe",
         "age":29,
         "emails":[
            "john.doe@gmail.com",
            "john.doe@pitt.edu"
         ],
         "is_ft":false
      },
      {
         "name":"Jane Doe",
         "age":21,
         "emails":[
            "jane.doe@gmail.com",
            "jane.doe@pitt.edu"
         ],
         "is_ft":true
      }
   ]
}
'''
```

### Load

```python
j_data = json.loads(emp_string)
j_data
```

```python
for emp in j_data["employees"]:
    print(emp)
```

```python
for emp in j_data["employees"]:
    print(emp['name'])
```

### Dump python object into a json string

```python
for emp in j_data['employees']:
    # del ---> removees key and its corresponding value
    del emp['age']

new_emp_str = json.dumps(j_data)

new_emp_str
```
